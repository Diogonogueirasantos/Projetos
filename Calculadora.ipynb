```python
from time import sleep
```

```python
class Calc:
    def __init__(self):
        self.inicializacaoCalc()
    
    def inicializacaoCalc(self):
        print( """Por favor a seguir selecione uma das operações abaixo:
                     [1] Adição
                     [2] Subtração
                     [3] Multiplação
                     [4] Divisão""")
        Opcao = int(input('selecione uma opção acima: '))

        match Opcao:
            case 1:
                self.Adicao()
            case 2:
                self.Subtracao()
            case 3:
                self.Multiplicacao()
            case 4:
                self.selecionador_Divisao()
                
        if Opcao > 4:
            print('Ops... parece que ocorreu um erro! \n', end=" ")
            print('   A opção selecionado não foi encontrada')
            return self.reinicializacaoCalc()

    def reinicializacaoCalc(self):
        print('Infelizmente a opção selecionada parece não existir...')
        Opcao = str(input('Deseja tentar novamente [S/N]? ')).strip().upper()
        match Opcao:
            case "S":
                self.inicializacaoCalc()
            case "N":
                print('Puxa é uma pena...')
                mensagem = """Aviso!
                              O programa irá se encerrar em alguns instantes."""
                print(mensagem)
                sleep(3)

    def Adicao(self):
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        Soma = N1 + N2
        print(f'A soma entre {N1} e {N2} é igual a {Soma}')
        Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()
        if Opcao in "S":
            self.inicializacaoCalc()
        else:
            print('Finalizando o programa...')
            sleep(2)

    def Subtracao(self):
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        Subtracao = N1 - N2
        print(f'A diferença entre {N1} e {N2} é igual a {Subtracao}')
        Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()
        if Opcao in "S":
            self.inicializacaoCalc()
        else:
            print('Finalizando o programa...')
            sleep(2)
    

    def Multiplicacao(self):
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        Produto = N1 * N2 
        print(f'O produto entre {N1} e {N2} é igual a {Produto}')
        Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()
        if Opcao in "S":
            self.inicializacaoCalc()
        else:
            print('Finalizando o programa...')
            sleep(2)


    def selecionador_Divisao(self):
        print("""Por favor selecione uma das opções abaixo
                      [1] Divisão por números inteiros
                      [2] Divisão por números reais
                      [3] Resto da Divisão""")
        Opcao = int(input('Selecione uma das opções acima: '))
        match Opcao:
            case 1:
                self.divisao_Inteiro()
            case 2:
                self.divisao_Real()
            case 3:
                self.resto_Divisao()


    def divisao_Inteiro(self):
        print("""Antes de iniciar o seu calculo é valido lembrar que para a este calculo ser realizado é necessário com que: a ∈ Z e b ≠ 0""")
        
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        if N2 == 0:
            print('Aviso! está operação não pode ser realizada por violar a condição citada acima')
            return self.divisao_Inteiro()
        else:
            Divisao = N1 // N2
            print(f'A divisão entre {N1} e {N2} é igual a {Divisao:.2f}')
            Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()
            
            if Opcao in "S":
                self.inicializacaoCalc()
            else:
                print('Finalizando o programa...')
                sleep(2)


    def divisao_Real(self):
        print("""Antes de iniciar o seu calculo é valido lembrar que para a este calculo ser realizado é necessário com que: a ∈ Z e b ≠ 0""")
        
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        if N2 == 0:
            print('Aviso! está operação não pode ser realizada por violar a condição citada acima')
            return self.divisao_Real()
        else:
            Divisao = N1 / N2
            print(f'A divisão entre {N1} e {N2} é igual a {Divisao:.2f}')
            Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()

            if Opcao in "S":
                self.inicializacaoCalc()
            else:
                print('Finalizando o programa...')
                sleep(2)


    def resto_Divisao(self):
        print("""Antes de iniciar o seu calculo é valido lembrar que para a este calculo ser realizado é necessário com que: a ∈ Z e b ≠ 0""")
        
        N1 = int(input('Digite um número: '))
        N2 = int(input('Digite outro: '))
        if N2 == 0:
            print('Aviso! está operação não pode ser realizada por violar a condição citada acima')
            return self.resto_Divisao()
        else:
            Resto = N1 % N2
            print(f'O resto da divisão é {Resto:.2f}')
            Opcao = str(input('Deseja continuar [S/N]?')).strip().upper()
            
            if Opcao in "S":
                self.inicializacaoCalc()
            else:
                print('Finalizando o programa...')
                sleep(2)
                
                
        
```
